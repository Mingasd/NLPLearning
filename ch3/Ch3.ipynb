## 3.4 使用正则表达式检测词组搭配

```python
import re
import nltk

wordlist = [w for w in nltk.corpus.gutenberg.words('austen-emma.txt') if w.islower()]
```

```python
# 基本的元字符
# $ 用来匹配单词的末尾
[w for w in wordlist if re.search('ed$',w)]
```

```python
# . 匹配任何单个字符
# ^ 匹配字符串的开始
set([w for w in wordlist if re.search('^..j..t..$',w)])
```

```python
# ？ 表示前面的字符是可选的
set([w for w in wordlist if re.search('^e-?mail$',w)])
```

```python
# 范围与闭包
set([w for w in wordlist if re.search('^[ghi][mno][jlk][def]$',w)])
```

## 3.5 正则表达式的应用

```python
word  = 'supercalifragilisticexpialidocious'
re.findall(r'[aeiou]',word)
```

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
fd = nltk.FreqDist(vs for word in wsj 
                   for vs in re.findall(r'[aeiou]{2,}',word))
fd.items()
```

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeios]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp,word)
    return "".join(pieces)
english_udhr = nltk.corpus.udhr.words('English-Latin1')
nltk.tokenwrap(compress(w) for w in english_udhr[:75])
```

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]',w)]
cfd =nltk.ConditionalFreqDist(cvs)
cfd.tabulate()
```

```python
cv_word_pairs = [(cv,w) for w in rotokas_words
                        for cv in re.findall(r'[ptksvr][aeiou]',w)]
cv_index = nltk.Index(cv_word_pairs)
cv_index['su']
```

```python
def stem(word):
    for suffix in ['ing','ly','ed','ious','ies','ive','es','s','ment']:
        if word.endswith(suffix):
            return word[:-len(suffix)]
    return word;
```

## 3.6 规范化文本

```python
raw = """DENNIS: Listen, strange women lying in ponds distributing swords is no basis for a system of government. Supreme executive power derives from a mandate from the masses, not from some farcical aquatic ceremony."""
tokens = nltk.word_tokenize(raw)
```

```python
porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()
[porter.stem(t) for t in tokens]
[lancaster.stem(t) for t in tokens]
```

```python
wnl = nltk.WordNetLemmatizer()
[wnl.lemmatize(t) for t in tokens]
```
