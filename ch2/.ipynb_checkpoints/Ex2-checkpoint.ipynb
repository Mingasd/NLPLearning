```python
# 4
from nltk.corpus import state_union
fileids = state_union.fileids()
modals = ['men','women','people']
text = state_union.words()
fdist = nltk.FreqDist(w.lower() for w in text)
for m in modals:
    print(m+':',fdist[m])

cfd = nltk.ConditionalFreqDist(
    (target,fileid[:4])
    for fileid in state_union.fileids()
    for w in state_union.words(fileid)
    for target in modals if w.lower().startswith(target))
cfd.plot()
```

```python
# 14
def supergloss(s):
    _str = ''
    for synset in s:
        _str = _str+synset.definition()
        for hypon in synset.hyponyms():
            _str = _str + hypon.definition()
        for hypern in synset.hypernyms():
            _str = _str + hypern.definition()
            
#         synset.definition()+synset.hyponyms()+synset.hypernyms()
    return _str
s = wn.synsets('car')
supergloss(s)
```

```python
# 16
genres = brown.categories()
for genre in genres:
    word_num = len([word for word in brown.words(categories=genre)])
    type_num = len(set(word for word in brown.words(categories=genre)))
    print(genre+':',word_num,type_num,word_num/type_num)
```

```python
# 22
def hedge(text):
    i = 0
    l = []
    for word in text:
        l.append(word)
        i = i+1
        if i%3==0:
            l.append('like')
    return l
hedge(sent2)
```

```python
# 25
from nltk.corpus import udhr
def find_language(s):
    langs = []
    for fileid in udhr.fileids():
        if fileid.endswith('-Latin1'):
            if s in  udhr.words(fileid):
                langs.append(fileid.replace('-Latin1',''))
    return langs

find_language('the')
```
