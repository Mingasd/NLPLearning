## 2.1 获得文本语料库

```python
from nltk.book import *
import nltk
```

```python
# 古腾堡语料库
from nltk.corpus import gutenberg
```

```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid)) 
    num_words = len(gutenberg.words(fileid)) 
    num_sents = len(gutenberg.sents(fileid)) 
    num_vocab = len(set([w.lower() for w in gutenberg.words(fileid)]))
    print(int(num_chars/num_words),int(num_words/num_sents),int(num_words/num_vocab),fileid)
    #平均词长 平均句子长度 本文中每个词出现的平均次数
```

```python
# 布朗语料库
from nltk.corpus import brown
```

```python
brown.categories()
brown.words(categories='news')
brown.words(fileids=['cg22'])
brown.sents(categories=['news','editorial','reviews'])
```

```python
news_text = brown.words(categories='news')
fdist = nltk.FreqDist([w.lower() for w in news_text])
modals = ['can','could','may','might','must','will']
for m in modals:
    print(m+':',fdist[m])
    
```

```python
cfd = nltk.ConditionalFreqDist(
    (genre,word)
    for genre in brown.categories()
    for word in brown.words(categories=genre))

genres = ['news','religion','hobbies','science_fiction','romance','humor']
modals = ['can','could','may','might','must','will']
cfd.tabulate(conditions = genres,samples=modals)
```

```python
# 路透社语料库
from nltk.corpus import reuters
reuters.fileids()
reuters.categories()
reuters.categories('training/9865')
```

```python
# 就职演说语料库
from nltk.corpus import inaugural

cfd = nltk.ConditionalFreqDist(
     (target,fileid[:4])
     for fileid in inaugural.fileids()
     for w in inaugural.words(fileid)
     for target in ['america','citizen']
     if w.lower().startswith(target))
cfd.plot()
```

## 2.2 条件频率分布

```python
genre_word = [(genre,word) 
              for genre in ['news','romance'] 
              for word in brown.words(categories=genre)]
len(genre_word)
```

```python
cfd = nltk.ConditionalFreqDist(genre_word)
cfd['news']
cfd['romance']['could']
```

```python
from nltk.corpus import udhr
languages = ['Chickasaw','English','German_Deutsch']
cfd = nltk.ConditionalFreqDist(
        (lang,len(word))
        for lang in languages
        for word in udhr.words(lang+'-Latin1'))
cfd.tabulate(conditions=['English','German_Deutsch'],samples=range(10),
            cumulative=True)
```

```python
days=['Monday', 'Tuesday', 'Wednesday','Thursday', 'Friday', 'Saturday', 'Sunday']
cfd = nltk.ConditionalFreqDist((c, d) for c in ['news','romance'] for d in brown.words(categories=c))
cfd.tabulate(samples=days)
cfd.plot(samples=days)
```

```python
sent = ['In','the','beginning','God','created','the','heaven'
       ,'and','the','earth','.']
for item in nltk.bigrams(sent):
    print(item)
```

```python
def generate_model(cfdist,word,num=15):
    for i in range(num):
        print(word)
        word = cfdist[word].max()
text = nltk.corpus.genesis.words('english-kjv.txt')
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)

generate_model(cfd,'living')
```

## 2.4 词典资源

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab.difference(english_vocab)
    return sorted(unusual)

unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))
unusual_words(nltk.corpus.nps_chat.words())
```

```python
names = nltk.corpus.names
names.fileids()
male_names = names.words('male.txt')
female_names = names.words('female.txt')
[w for w in male_names if w in female_names]
```

```python
cfd = nltk.ConditionalFreqDist(
        (fileid,name[-1])
        for fileid in names.fileids()
        for name in names.words(fileid))
cfd.plot()
```

```python
entries = nltk.corpus.cmudict.entries()
for entry in entries[39943:39951]:
    print(entry)
```

```python
from nltk.corpus import swadesh
swadesh.fileids()
swadesh.words('en')
fr2en = swadesh.entries(['fr','en'])
translate = dict(fr2en)
translate['chien']
de2en = swadesh.entries(['de','en'])
es2en = swadesh.entries(['es','en'])
translate.update(dict(de2en))
translate.update(dict(es2en))
translate['Hund']
translate['perro']
```

## 2.5 WordNet

```python
# WordNet是面向语义的英语词典
from nltk.corpus import wordnet as wn
```

```python
wn.synsets('motorcar')
wn.synset('car.n.01').lemma_names()
wn.synset('car.n.01').definition()
wn.synset('car.n.01').examples()
wn.synset('car.n.01').lemmas()
```

```python
#得到指定同义词集的所有词条
wn.synset('car.n.01').lemmas()
#查找特定的词条
wn.lemma('car.n.01.automobile')
#得到一个词条对应的同义词集
wn.lemma('car.n.01.automobile').synset()
#得到一个词条的名字
wn.lemma('car.n.01.automobile').name()
```

```python
motorcar = wn.synset('car.n.01')
types_of_motorcar = motorcar.hyponyms()
types_of_motorcar[26]
sorted([lemma.name() for synset in types_of_motorcar for lemma in synset.lemmas()])
```

```python
motorcar.hypernyms()
paths = motorcar.hypernym_paths()
len(paths)
[synset.name for synset in paths[0]]
[synset.name for synset in paths[1]]
motorcar.root_hypernyms()
```

```python
# 语义相似度
right = wn.synset('right_whale.n.01')
orca = wn.synset('orca.n.01')
minke = wn.synset('minke_whale.n.01')
tortoise = wn.synset('tortoise.n.01')
novel = wn.synset('novel.n.01')
right.lowest_common_hypernyms(minke)
right.lowest_common_hypernyms(orca)
right.lowest_common_hypernyms(tortoise)
right.lowest_common_hypernyms(novel)
wn.synset('baleen_whale.n.01').min_depth()
right.path_similarity(minke)
```
