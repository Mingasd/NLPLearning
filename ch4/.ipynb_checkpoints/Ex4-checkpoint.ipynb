```python
# 14
import nltk
words = nltk.corpus.gutenberg.words('austen-emma.txt')

def novel10(text):
    l = len(text)
    words1 = text[:int(l*9/10)]
    words2 = text[int(-l/10):]
    for w in words2:
        if w not in words1:
            print(w)

novel10(words)
```

```python
# 15
sent = ['the','dog','gave','john','the','food','dog','food','a']
dictFreq = {}
for w in sent:
    dictFreq[w] = dictFreq.get(w,0)+1
dictFreq = sorted(dictFreq.items(),key = lambda k : k[0])
for k, v in dictFreq:
    print(k, v)
```

```python
# 16
letter_vals = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5, 'f':80, 'g':3, 'h':8,'i':10, 'j':10, 'k':20, 'l':30, 'm':40, 'n':50, 'o':70, 'p':80, 'q':100,'r':200, 's':300, 't':400, 'u':6, 'v':6, 'w':800, 'x':60, 'y':10, 'z':7}

def gematria(word):
    num = 0
    for c in word:
            num = num + letter_vals[c]
    return num
text = nltk.corpus.state_union.words()[:1000]
dic = {}
for w in text:
    if w.isalpha():
        w = w.lower()
        dic[w] = dic.get(w,0)+ gematria(w)

for k,v in dic.items():
    print(k,v)
```

```python
# 17

def shorten(text,n):
    s = sorted(nltk.FreqDist(text).items(),key =lambda k :k[1],reverse=True)
    text = " ".join(text)
    for i in range(n):
        text = text.replace(s[i][0],'')
    text = text.split()
    return text
shorten(text,5)
```
