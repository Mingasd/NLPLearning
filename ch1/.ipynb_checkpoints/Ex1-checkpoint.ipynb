```python
# 17
index =  text9.index('sunset')
start = 0
end = len(text9)
for i in range(index,0,-1):
    if text9[i]=='!' or text9[i]=='.' or text9[i]=='?':
        start = i;
        break;

for i in range(index,end):
    if text9[i]=='.' or text9[i]=='!'or text9[i]=='?':
        end = i;
        break;

mySent = text9[start+1:end+1]
mySent
```
